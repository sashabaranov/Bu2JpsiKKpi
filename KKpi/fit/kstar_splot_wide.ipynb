```python
import ROOT, random

from Ostap import *
from Ostap.PyRoUts import *
from Ostap.Selectors import *
import Ostap.FitModels as Models
from MyFitModels import Charm3_pdf

from tools import *
from data import selection7

from ostap import *
from mc_average import *

def variate(var, minval=None, maxval=None, startval=None):
    if minval: var.setMin(minval)
    if maxval: var.setMax(maxval)
    if startval: var.setVal(startval)
```

```python
#
resonant = [
    ( 'DTFm_kpi', "Inv. \, Mass(K\pi), GeV/c^2", 0.65, 1.55),
]

# Bu meson
m_Bu = ROOT.RooRealVar('DTFm_b', 'mass(J/psiKKpi)', 5.22, 5.34)
nbin_Bu = 30
binning_b = (m_Bu.getMax() - m_Bu.getMin()) / nbin_Bu * 1000


# K*(892) mass
m_Kstar = ROOT.RooRealVar('DTFm_kpi', 'mass(K \pi)', 0.65, 1.55)
nbin_Kstar = 45
events_binning = (m_Kstar.getMax() - m_Kstar.getMin()) / nbin_Kstar * 1000

# For Breit-Wigner
mean_Kstar = ROOT.RooRealVar('mean_Kstar', 'mean(K*)', 0.85, 0.9)
mean_Kstar.fix(0.892)

width_Kstar = ROOT.RooRealVar('width_Kstar', 'width(Kstar0)', 0.0, 1.0)

# PhaseSpace range
low_kpi = ROOT.RooRealVar('low', 'low_KstarBg', 0.6, 0.7)
low_kpi.fix(0.634)

high_kpi = ROOT.RooRealVar('high', 'high_KstarBg', 1.3, 2.0)





selector_variables = [(m_Bu, lambda s: s.DTFm_b ), (m_Kstar, lambda s: s.DTFm_kpi )] + resonant
```

Basically, select all needed variables:

```python
# Trigger
cuts_trg =  "((psi_l0tos & 2) == 2)"
cuts_trg += "&& ((psi_l1tos & 2) == 2)"
cuts_trg += "&& ((psi_l2tos & 2) == 2)"

# Cuts
cuts_ = " DTFchi2ndof > 0"
cuts_ += "&& DTFchi2ndof < 5"
cuts_ += "&& DTFctau > 0.25"
cuts_ += "&& vchi2_b < 20"
cuts_ += "&& pt_kaon[0] > 0.6 && pt_kaon[1] > 0.6"
cuts_ += "&& pt_pion > 0.3"
cuts_ += "&& m_jpsi    > 3.020 && m_jpsi    < 3.135"
cuts_ += "&& minann_K  > 0.3"
cuts_ += "&& ann_pion[0] > 0.3"
cuts_ += "&& MIPCHI2DV_k1 > 12. && MIPCHI2DV_k2 > 12. && MIPCHI2DV_pi > 12."

cuts_ += "&& " + cuts_trg
```

```python
sel_Bu = SelectorWithVarsCached(
    variables=selector_variables,
    selection=cuts_,
    files=selection7.files
)

if not sel_Bu._loaded_from_cache:
    selection7.chain.process(sel_Bu)
ds_Bu = sel_Bu.dataset()
ds_Bu.Print('v')
```

```python
canvas = rootnotes.random_canvas()

frame = m_Bu.frame(nbin_Bu)
ds_Bu.plotOn(frame)
frame.Draw()

canvas
```

```python
s1_Bu = Models.CB2_pdf(
    'Bu1',
    m_Bu.getMin(),
    m_Bu.getMax(),
    fixMass=5.28,
    fixSigma=5.75e-3,
    fixAlphaL=avg_mc("aL_Bu1"),
    fixAlphaR=avg_mc("aR_Bu1"),
    fixNL=avg_mc("nL_Bu1"),
    fixNR=avg_mc("nR_Bu1"),
    mass=m_Bu
)

bkg_Bu = Models.Bkg_pdf('BBu', mass=m_Bu, power=1)

# model_B = Models.Fit1D(
#     signal=s1_Bu,
#     background=bkg_Bu,
#     components=[kkk]
# )


model_B = Charm3_pdf(
    signal=s1_Bu,
    background=bkg_Bu,
)


# bkg_Bu.tau.fix(0)
# bkg_Bu.phis[0].fix(0)
```

```python
c1 = rootnotes.random_canvas(size=(1024, 768))


ru, fu = model_B.fitTo(ds_Bu, draw=True, nbins=nbin_Bu)

model_B.signal.mean.release()
model_B.signal.sigma.release()

ru, fu = model_B.fitTo(ds_Bu, draw=True, nbins=nbin_Bu)


fu.SetTitle("")
fu.SetXTitle('Inv.\,mass(J/\psi KK\pi), GeV/c^{2}}')
fu.SetYTitle("Events / %.1f MeV/c^{2}" % binning_b)

fu.Draw()
model_B.legend.Draw()

c1
```

### Now fit $K^{*}$

```python
hist = ROOT.TH1F(hID(), '', nbin_Kstar, m_Kstar.getMin(), m_Kstar.getMax())
hist.Sumw2()

ds_Bu.project(hist, "DTFm_kpi", "S_sw")
```

```python
canvas = rootnotes.random_canvas(size=(1024, 768))
hist.Draw()
canvas
```

```python
m = VE(0.8916, 0.02**2)
bw = cpp.Gaudi.Math.BreitWigner(m.value(), m.error(), 0.493, 0.139, 1, 1)


s_Ks =  Models.BreitWigner_pdf(
   "Ks1",
   m_Kstar.getMin(),
   m_Kstar.getMin(),
   bw,
   mass=m_Kstar,
   fixMass=0.8958,
   fixGamma=47.4e-3,
#    convolution=1e-3
)



ps = cpp.Gaudi.Math.PhaseSpaceNL(low_kpi, high_kpi, 2, 4) 


model_Ks = Models.Fit1D(
    signal=s_Ks,
    background=Models.PSPol_pdf( 'Bkg' , m_Kstar, ps , 0 ),
    suffix='Ks'
)
```

```python
canvas = rootnotes.random_canvas(size=(1024, 768))

r, f = model_Ks.fitHisto(hist, draw=True)

# model_Ks.signal.gamma.release()
# r, f = model_Ks.fitHisto(hist, draw=True)

# model_Ks.signal.mean.release()
# r, f = model_Ks.fitHisto(hist, draw=True)


f.SetTitle("")
f.SetXTitle("Inv.\,mass(K\pi), GeV/c^{2}}")
f.SetYTitle("Events / %.1f MeV/c^{2}" % events_binning)



canvas
```
