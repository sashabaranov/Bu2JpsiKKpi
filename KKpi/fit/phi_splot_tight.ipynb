```python
import ROOT, random

from Ostap import *
from Ostap.PyRoUts import *
from Ostap.Selectors import *
import Ostap.FitModels as Models
from MyFitModels import Charm3_pdf

from tools import *
from data import selection7

from ostap import *
from mc_average import *


def variate(var, minval=None, maxval=None, startval=None):
    if minval: var.setMin(minval)
    if maxval: var.setMax(maxval)
    if startval: var.setVal(startval)
```

```python
#
resonant = [
    ( 'DTFm_kk', "Inv. \, Mass(KK), GeV/c^2", 1.0, 1.06),
]

# Bu meson
m_Bu = ROOT.RooRealVar('DTFm_b', 'mass(J/psiKKpi)', 5.22, 5.34)
nbin_Bu = 30
binning_b = (m_Bu.getMax() - m_Bu.getMin()) / nbin_Bu * 1000


# Phi mass
m_Phi = ROOT.RooRealVar('DTFm_kk', 'mass(KK)', 1.0, 1.06)
nbin_Phi = 20
events_binning = (m_Phi.getMax() - m_Phi.getMin()) / nbin_Phi * 1000


# For Breit-Wigner
mean_Phi = ROOT.RooRealVar('mean_Phi', 'mean(Phi)', 1.0, 1.05)
mean_Phi.fix(1.02)

width_Phi = ROOT.RooRealVar('width_Phi', 'width(Phi)', 0.0, 1.0)

# PhaseSpace range
low_kk = ROOT.RooRealVar('low', 'low_PhiBg', 0.9, 1.0)
low_kk.fix(0.986)

high_kk = ROOT.RooRealVar('high', 'high_PhiBg', 2.0, 2.1)





selector_variables = [(m_Bu, lambda s: s.DTFm_b )] + resonant
```

Basically, select all needed variables:

```python
# Trigger
cuts_trg =  "((psi_l0tos & 2) == 2)"
cuts_trg += "&& ((psi_l1tos & 2) == 2)"
cuts_trg += "&& ((psi_l2tos & 2) == 2)"

# Cuts
cuts_ = " DTFchi2ndof > 0"
cuts_ += "&& DTFchi2ndof < 5"
cuts_ += "&& DTFctau > 0.25"
cuts_ += "&& vchi2_b < 20"
cuts_ += "&& pt_kaon[0] > 0.6 && pt_kaon[1] > 0.6"
cuts_ += "&& pt_pion > 0.3"
cuts_ += "&& m_jpsi    > 3.020 && m_jpsi    < 3.135"
cuts_ += "&& minann_K  > 0.3"
cuts_ += "&& ann_pion[0] > 0.3"
cuts_ += "&& MIPCHI2DV_k1 > 12. && MIPCHI2DV_k2 > 12. && MIPCHI2DV_pi > 12."

cuts_ += "&& " + cuts_trg
```

```python
sel_Bu = SelectorWithVarsCached(
    variables=selector_variables,
    selection=cuts_,
    files=selection7.files
)

if not sel_Bu._loaded_from_cache:
    selection7.chain.process(sel_Bu)
ds_Bu = sel_Bu.dataset()
ds_Bu.Print('v')
```

```python
canvas = rootnotes.random_canvas()

frame = m_Bu.frame(nbin_Bu)
ds_Bu.plotOn(frame)
frame.Draw()

canvas
```

```python
s1_Bu = Models.CB2_pdf(
    'Bu1',
    m_Bu.getMin(),
    m_Bu.getMax(),
    fixMass=5.28,
    fixSigma=5.75e-3,
    fixAlphaL=avg_mc("aL_Bu1"),
    fixAlphaR=avg_mc("aR_Bu1"),
    fixNL=avg_mc("nL_Bu1"),
    fixNR=avg_mc("nR_Bu1"),
    mass=m_Bu
)

bkg_Bu = Models.Bkg_pdf('BBu', mass=m_Bu, power=1)

# model_B = Models.Fit1D(
#     signal=s1_Bu,
#     background=bkg_Bu,
#     components=[kkk]
# )


model_B = Charm3_pdf(
    signal=s1_Bu,
    background=bkg_Bu,
)


# bkg_Bu.tau.fix(0)
# bkg_Bu.phis[0].fix(0)
```

```python
c1 = rootnotes.random_canvas(size=(1024, 768))


ru, fu = model_B.fitTo(ds_Bu, draw=True, nbins=nbin_Bu)

model_B.signal.mean.release()
model_B.signal.sigma.release()

ru, fu = model_B.fitTo(ds_Bu, draw=True, nbins=nbin_Bu)


fu.SetTitle("")
fu.SetXTitle('Inv. mass(J/#psi KK#pi), GeV/c^{2}')
fu.SetYTitle("Events / %.1f MeV/c^{2}" % binning_b)

model_B.legend.AddEntry("", "N_{sig} = 21 #pm 6", "")

fu.Draw()
model_B.legend.Draw()
model_B.legend.Draw()

c1
```

### Now fit $\phi$

```python
hist = ROOT.TH1F(hID(), '', nbin_Phi, m_Phi.getMin(), m_Phi.getMax())
hist.Sumw2()

ds_Bu.project(hist, "DTFm_kk", "S_sw")
```

```python
canvas = rootnotes.random_canvas(size=(1024, 768))
hist.Draw()
canvas
```

```python
bw     = cpp.Gaudi.Math.Phi0 ( 1.0195 ,
                               0.0043 ,
                               0.4937 )

s_Phi =  Models.BreitWigner_pdf(
   "Phi1",
   m_Phi.getMin(),
   m_Phi.getMin(),
   bw,
   mass=m_Phi,
#    convolution = 2e-3,
   fixMass = 1.0195,
   fixGamma = 4.3e-3
)


ps2    = cpp.Gaudi.Math.PhaseSpaceNL ( 2*0.4937 , 5.278-3.096 , 2 , 4 ) 

model_Phi = Models.Fit1D(
    signal=s_Phi,
    background=Models.PSPol_pdf( 'Bkg' , m_Phi , ps2 , 0 ),
    suffix='Phi'
)
```

```python
canvas = rootnotes.random_canvas(size=(1024, 768))

r, f = model_Phi.fitHisto(hist, draw=True)


# model_Phi.signal.mean.release()
# r, f = model_Phi.fitHisto(hist, draw=True)

f.SetTitle('')
f.SetXTitle("Inv.\,mass(KK), GeV/c^{2}}")
f.SetYTitle("Events / %.1f MeV/c^{2}" % events_binning)

legend = ROOT.TLegend(0.55, 0.8, 0.9, 0.9)
legend.SetFillColor(ROOT.kWhite)
# legend.SetTextSize(0.04)

legend.AddEntry(ds_Bu, "N_{\phi} = 19 \pm 5", "")
legend.Draw()


canvas
```
