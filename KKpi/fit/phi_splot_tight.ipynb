```python
import ROOT, random

from Ostap import *
from Ostap.PyRoUts import *
from Ostap.Selectors import *
import Ostap.FitModels as Models
from MyFitModels import Charm3_pdf

from tools import *
from data import selection7

from ostap import *

def variate(var, minval=None, maxval=None, startval=None):
    if minval: var.setMin(minval)
    if maxval: var.setMax(maxval)
    if startval: var.setVal(startval)
```

```python
# Bu meson
m_Bu = ROOT.RooRealVar('DTFm_b', 'mass(J/psiKKpi)', 5.22, 5.42)
import shelve
db = shelve.open("$KKpidir/fit/histos.shelve")

# kkk_hist = db['KKK']['MC']['p8_k3_cuts']
kpipi_hist = db['Kpipi']['MC']['p8_pi1']
# kkk = Models.H1D_pdf(name=kkk_hist.GetName(), mass=m_Bu, histo=smear_kkk(kkk_hist))
kpipi = Models.H1D_pdf(name=kpipi_hist.GetName(), mass=m_Bu, histo=smear_kpipi(kpipi_hist))
```

```python
#
resonant = [
    ( 'DTFm_kk', "Inv. \, Mass(KK), GeV/c^2", 0.0, 20.0),
    ( 'DTFm_kpi', "Inv. \, Mass(K\pi), GeV/c^2", 0.0, 20.0),
]

# Bu meson
m_Bu = ROOT.RooRealVar('DTFm_b', 'mass(J/psiKKpi)', 5.22, 5.42)
nbin_Bu = 25
binning_b = (m_Bu.getMax() - m_Bu.getMin()) / nbin_Bu * 1000


# Phi mass
m_Phi = ROOT.RooRealVar('DTFm_kk', 'mass(KK)', 1.0, 1.1)
nbin_Phi = 20
events_binning = (m_Phi.getMax() - m_Phi.getMin()) / nbin_Phi * 1000


# For Breit-Wigner
mean_Phi = ROOT.RooRealVar('mean_Phi', 'mean(Phi)', 1.0, 1.05)
mean_Phi.fix(1.02)

width_Phi = ROOT.RooRealVar('width_Phi', 'width(Phi)', 0.0, 1.0)

# PhaseSpace range
low_kk = ROOT.RooRealVar('low', 'low_PhiBg', 0.9, 1.0)
low_kk.fix(0.986)

high_kk = ROOT.RooRealVar('high', 'high_PhiBg', 2.0, 2.1)





selector_variables = [(m_Bu, lambda s: s.DTFm_b ), (m_Phi, lambda s: s.DTFm_kk )] + resonant
```

Basically, select all needed variables:

```python
# Trigger
cuts_trg =  "((psi_l0tos & 2) == 2)"
cuts_trg += "&& ((psi_l1tos & 2) == 2)"
cuts_trg += "&& ((psi_l2tos & 2) == 2)"

# Cuts
cuts_ = " DTFchi2ndof > 0"
cuts_ += "&& DTFchi2ndof < 5"
cuts_ += "&& DTFctau > 0.25"
cuts_ += "&& vchi2_b < 20"
cuts_ += "&& pt_kaon[0] > 0.6 && pt_kaon[1] > 0.6"
cuts_ += "&& pt_pion > 0.3"
cuts_ += "&& m_jpsi    > 3.020 && m_jpsi    < 3.135"
cuts_ += "&& minann_K  > 0.3"
cuts_ += "&& ann_pion[0] > 0.3"
cuts_ += "&& MIPCHI2DV_k1 > 12. && MIPCHI2DV_k2 > 12. && MIPCHI2DV_pi > 12."

cuts_ += "&& " + cuts_trg
```

```python
sel_Bu = SelectorWithVarsCached(
    variables=selector_variables,
    selection=cuts_,
    files=selection7.files
)

if not sel_Bu._loaded_from_cache:
    selection7.chain.process(sel_Bu)
ds_Bu = sel_Bu.dataset()
ds_Bu.Print('v')
```

```python
canvas = rootnotes.random_canvas()

frame = m_Bu.frame(nbin_Bu)
ds_Bu.plotOn(frame)
frame.Draw()

canvas
```

```python
s1_Bu = Models.CB2_pdf(
    'Bu1',
    m_Bu.getMin(),
    m_Bu.getMax(),
    fixMass=5.2792e+0,
    fixSigma=0.006,
    fixAlphaL=2.1018e+00,
    fixAlphaR=1.9818e+00,
    fixNL=6.1464e-01,
    fixNR=2.1291e+00,
    mass=m_Bu
)

bkg_Bu = Models.Bkg_pdf('BBu', mass=m_Bu, power=1)

# model_B = Models.Fit1D(
#     signal=s1_Bu,
#     background=bkg_Bu,
#     components=[kkk]
# )


model_B = Charm3_pdf(
    signal=s1_Bu,
    signal2=kpipi,
    background=bkg_Bu,
)


# bkg_Bu.tau.fix(0)
# bkg_Bu.phis[0].fix(0)
```

### Now fit $\phi$

```python
hist = ROOT.TH1F(hID(), '', nbin_Phi, m_Phi.getMin(), m_Phi.getMax())
hist.Sumw2()

ds_Bu.project(hist, "DTFm_kk", "S_sw")
```

```python
canvas = rootnotes.random_canvas(size=(1024, 768))
hist.Draw()
canvas
```

```python
m = VE(1.02, 0.02**2)
bw = cpp.Gaudi.Math.BreitWigner(m.value(), m.error(), 0.493, 0.493, 1, 1)


s_Phi =  Models.BreitWigner_pdf(
   "Phi1",
   m_Phi.getMin(),
   m_Phi.getMin(),
   bw,
   mass=m_Phi,
   mean        = mean_Phi , 
   gamma       = width_Phi)


ps = cpp.Gaudi.Math.PhaseSpaceNL(low_kk, high_kk, 2, 4) 

model_Phi = Models.Fit1D(
    signal=s_Phi,
    #background = Models.Bkg_pdf ( 'BPhi' , mass = m_Phi , power = 4 ) ,
    background=Models.PSPol_pdf( 'Bkg' , m_Phi , ps , 1),
#     background=Models.PhaseSpace_pdf(
#         'BPhi', x=m_Phi, low=low_kk, high=high_kk, N=2, L=4, power=0),
    suffix='Phi'
)

model_Phi.signal.mean.fix(1.020)
model_Phi.signal.gamma.fix(4.26e-3)
```
